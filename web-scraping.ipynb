```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
BASE_URL = 'https://habr.com'
```

```python
resp = requests.get(BASE_URL + '/ru/search/', params={
    'q': 'python',
    'target_type': 'posts',
    'order': 'relevance'
})
```

```python
resp.status_code
```

```python
soup = BeautifulSoup(resp.text)
```

```python
print(soup.prettify())
```

```python
results = []

for article_tag in soup.find_all('article'):
    datetime = article_tag.find('time').get('datetime')

    h2_tag = article_tag.find('h2')
    title = h2_tag.get_text()
    link = h2_tag.find('a').get('href')

    rating = article_tag.find('span', class_='tm-votes-meter__value').get_text()

    results.append([datetime, title, link, rating])

X = pd.DataFrame(results, columns=['Date', 'Title', 'Link', 'Rating'])
```

```python
X.head()
```

```python
X.Link = BASE_URL + X.Link
```

```python
X.Date = pd.to_datetime(X.Date).dt.date
```
